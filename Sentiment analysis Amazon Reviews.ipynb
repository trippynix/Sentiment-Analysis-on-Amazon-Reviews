```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df
```

```python
# The column names are actually a row let's extract it and change the column names and put it into the the actual data
newRow = pd.DataFrame([df.columns.tolist()], columns = df.columns)
data = pd.concat([newRow, df], ignore_index = True)

# Renaming the data columns
data.columns = ['Polarity', 'Title', 'Text']
data
```

### The Polarity indicates negative and positive reviews (1 == Negative, 2 == Positive)
### The Title column is title of the comment and Text column is the actual comment

```python
# Check for null values
print('Null data: \n', data.isnull().sum())
print('The null data in dataframe: \n')
data[data.isnull().any(axis=1)]
```

```python
# dropping all null value rows
data.dropna(inplace = True)
```

```python
# checking for maximum and minimum number of words in a single comment or Text column
mini = 10000
maxi = 0
for i in range(len(data)):
    mini = min(mini, len(list(data.iloc[i]['Text'].split())))
    maxi = max(maxi, len(list(data.iloc[i]['Text'].split())))
print(maxi)
print(mini)

# printing average number of words
print(round((maxi+mini)/2))
```

```python
# Extracting features and Target
train_features = data[['Text']]
train_target = data[['Polarity']]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import ComplementNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline


MultinomialNB = Pipeline(
    [
        ("vect", TfidfVectorizer()),
        ("clf", MultinomialNB()),
    ]
)

MultinomialNB.fit(train_features, train_target)
```
